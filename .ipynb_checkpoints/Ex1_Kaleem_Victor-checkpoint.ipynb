```python
import brian2 as b2
import neurodynex.exponential_integrate_fire.exp_IF as exp_IF

#Set default rheobase threshold
exp_IF.RHEOBASE_THRESHOLD_v_rh = -57*b2.mV

I_rh = (exp_IF.RHEOBASE_THRESHOLD_v_rh - exp_IF.V_REST - exp_IF.SHARPNESS_delta_T)/exp_IF.MEMBRANE_RESISTANCE_R
print("The minimal amplitude necessary for a constant current is:")
I_rh
```

```python
% matplotlib inline
import brian2 as b2
import neurodynex.exponential_integrate_fire.exp_IF as exp_IF
from neurodynex.tools import plot_tools, input_factory
rcParams['figure.figsize'] = 10, 8

input_current = input_factory.get_step_current(
    t_start=10, t_end=310, unit_time=b2.ms, amplitude=I_rh)

state_monitor, spike_monitor = exp_IF.simulate_exponential_IF_neuron(
    I_stim=input_current, simulation_time=300*b2.ms,
    v_spike=+15 * b2.mV,
    v_rheobase=-57 * b2.mV)

plot_tools.plot_voltage_and_current_traces(
    state_monitor, input_current,title="step current", 
    firing_threshold=exp_IF.FIRING_THRESHOLD_v_spike)
print("nr of spikes: {}".format(spike_monitor.count[0]))
```

#### explanation

```python
delta = 4 * b2.pamp
I_rh_delta = I_rh + delta

input_current = input_factory.get_step_current(
    t_start=10, t_end=310, unit_time=b2.ms, amplitude=I_rh_delta)

state_monitor, spike_monitor = exp_IF.simulate_exponential_IF_neuron(
    I_stim=input_current, simulation_time=300*b2.ms,
    v_spike=+15 * b2.mV,
    v_rheobase=-57 * b2.mV)

plot_tools.plot_voltage_and_current_traces(
    state_monitor, input_current,title="step current", 
    firing_threshold=exp_IF.FIRING_THRESHOLD_v_spike)
print("nr of spikes: {}".format(spike_monitor.count[0]))
```

## 3.2 Strength - Duration Curve [3+2 points]

The minimal amplitude to elicit a spike depends on the duration of the current. For an infinitely long current, we’ve just calculated the rheobase current. For short pulses and step currents, we can “experimentally” determine the minimal currents. If we plot the amplitude versus duration, we get the strength-duration curve.

At the end of the script, the strength-duration curve is plotted. Discuss it. You may want to add a log-log plot to better see the asymptotic behaviour.

```python
% matplotlib inline
import brian2 as b2
import matplotlib.pyplot as plt
import neurodynex.exponential_integrate_fire.exp_IF as exp_IF
from neurodynex.tools import plot_tools, input_factory
rcParams['figure.figsize'] = 10, 8

i=2 #change i and find the value that goes into min_amp
durations = [1,   2,    5,  10,   20,   50, 100]
min_amp =   [7.08, 3.64, 1.58, 0.89, 0.56, 0.38,0.33]

t=durations[i]
I_amp = min_amp[i]*b2.namp
title_txt = "I_amp={}, t={}".format(I_amp, t*b2.ms)

input_current = input_factory.get_step_current(t_start=10, t_end=10+t-1, unit_time=b2.ms, amplitude=I_amp)

state_monitor, spike_monitor = exp_IF.simulate_exponential_IF_neuron(I_stim=input_current, simulation_time=(t+25)*b2.ms,
                                                                     v_rheobase=-57 * b2.mV)

plot_tools.plot_voltage_and_current_traces(state_monitor, input_current,
                                           title=title_txt, firing_threshold=exp_IF.FIRING_THRESHOLD_v_spike,
                                          legend_location=2)
print("nr of spikes: {}".format(spike_monitor.count[0]))

plt.figure()
plt.plot(durations, min_amp)
plt.title("Strength-Duration curve")
plt.xlabel("t [ms]")
plt.ylabel("min amplitude [nAmp]")

plt.figure()
plt.loglog(durations, min_amp)
plt.title("Logarithmic Strength-Duration curve")
plt.xlabel("t [ms]")
plt.ylabel("min amplitude [nAmp]")
```

We see that the amplitude of the injected current necessary to elicit a spike decays exponentially with respect to time. The sterngth-duration curve, approaching the rheobase current in the horizontal asymptote.

First, try to get some intuition on shape of nullclines by plotting or simply sketching them on a piece of paper and answering the following questions.
Plot or sketch the u- and w- nullclines of the AdEx model (`I(t) = 0`)
- How do the nullclines change with respect to `a`?
- How do the nullclines change if a constant current `I(t) = c` is applied?
- What is the interpretation of parameter `b`?
- How do flow arrows change as tau_w gets bigger?

[max 10 lines, 4 points]

#### your answer
Changing *a* changes the slope of the w-nullcline. a=0 implies a horizontal nullcline with values further away creating a steeper slope. The u-nullcline is not affected.

Applying a constant current results in a vertical shift of the u-nullcline in proportion to the magnitude of the current applied. The w-nullcline is not affected.

The b parameter controls the "jump amplitude" of w, the spike triggered adaption that occurs after each spike. In the absence of the a,b parameters, the model no longer reflects adaption and returns to a tonic firing pattern.

As tau_w increases, du/dt decreases which causes the vertical flow component to decrease. The flow arrows thus becomes more horizontal at all points except at the u-nullcline, where the horizontal component is 0. The length of the flow arrows (magnitude) gets smaller as it approaches w-nullcline and flips as it crosses it.

As tau_w gets bigger, in particular relative to the value of tau_m, the vertical component of the flow arrows decreases at all points. This causes the resultant flow arrows to become more horizontal at all points except the u-nullcline where the horizontal component is 0.

#### your answer

*a* being small coupled with a small tau_w and b will lead to minimal adaption, and as expected the response is tonic with little inhibition being applied to the membrane potential. In contrast, keeping tau_w and b big creates a negative driving current which leads to a longer refractory period between spikes.

To conclude, there are several different possible firing patterns when a is small, common to them all is that the after spike adaptation will be the most significant adaptation taking place.
