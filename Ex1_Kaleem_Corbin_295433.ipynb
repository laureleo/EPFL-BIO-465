your answer

The a parameter acts to control the sensitivity of the adaption current to voltage even without spikes (i.e. subthreshold behavior). The V-nullcline remains unchanged with respect to a, however the w-nullcline is a straight line and will have its slope change with respect to a.

A change of the current by a constant value results in a vertical shift of the V-nullcline without changing its shape. As the current increases, the V-nullcline shifts upwards; the w-nullcline remains unchanged with respect to the current. 

The b parameter is used to characterize spike-triggered adapation, the adapation current w can will be increased by an amount b after a spike compared to right before the spike. In the absence of the a,b parameters, the model no longer reflects adaption and returns to a tonic firing pattern.

As tau_w gets bigger, in particular relative to the value of tau_m, the vertical component of the flow arrows (w-component) becomes insignificant relative to the horizontal component (v-component). Therefore, outside of the v-nullcline, the flow arrows are virtually horizontal, whereas the only vertical flow arrows exist along the w-nuclline. The length of the flow arrows (magnitude) gets smaller as it approaches w-nullcline and flips as it crosses it.
